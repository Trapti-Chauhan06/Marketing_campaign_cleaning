```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv", sep="\t")
df.head()
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.strip()
      .str.replace(" ", "_")
)
```

```python
df.dtypes
```

```python
df["education"].unique()
df["marital_status"].unique()
```

```python
df["education"] = df["education"].replace({
    "2n Cycle": "Master",
    "Basic": "High School",
    "Absurd": "Other",
    "YOLO": "Other"
})
df["marital_status"] = df["marital_status"].replace({
    "Alone": "Single",
    "Absurd": "Other",
    "YOLO": "Other"
})
```

```python
df.shape
```

```python
df.describe
```

```python
df["dt_customer"].dtype
```

```python
df["dt_customer"] = pd.to_datetime(
    df["dt_customer"],
    format="%d-%m-%Y"
)
```

```python
df["dt_customer"].dtype
```

```python
df = df[(df["year_birth"] > 1900) & (df["year_birth"] < 2000)]
upper_limit = df["income"].quantile(0.99)
df.loc[df["income"] > upper_limit, "income"] = upper_limit
```

```python
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
